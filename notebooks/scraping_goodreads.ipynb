```python
import pandas as pd

# Carregar os dados
df = pd.read_csv('../data/top_books.csv')

# Exibir as primeiras linhas do dataframe
print(df.head())

# Análise básica
print(df.describe())

# Visualização
df['Rating Value'] = df['Rating'].str.extract(r'(\d\.\d+)').astype(float)
df.plot(kind='hist', y='Rating Value', bins=20, title='Distribuição de Avaliações')
```
